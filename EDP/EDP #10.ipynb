```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def gnerate_vel_and_ac(exp, var, T, V0, numb):
    arr = np.zeros(numb)
    a = np.random.normal(exp, var ** 0.5, numb)
    arr[0] = V0
    for i in range(1, len(arr)):
        arr[i] = arr[i - 1] + a[i - 1] * T
    return arr, a
def generate_x(T, x0, a, v):
    arr = np.zeros(len(a))
    arr[0] = x0
    for i in range(1, len(a)):
        arr[i] = arr[i - 1] + v[i - 1] * T + a[i - 1] * (T ** 2) / 2
    return arr
def generate_trajectory(exp, var, T, x0, vx0, y0, vy0, numb):
    vx, ax = gnerate_vel_and_ac(exp, var, T, vx0, numb)
    x = generate_x(T, x0, ax, vx)
    
    vy, ay = gnerate_vel_and_ac(exp, var, T, vy0, numb)
    y = generate_x(T, y0, ay, vy)
    
    X = np.zeros([numb, 4])
    X[:, 0] = x
    X[:, 1] = vx
    X[:, 2] = y
    X[:, 3] = vy
    
    return X

def generate_D(x, y):
    D = np.zeros([len(x), 2])
    for i in range(len(x)):
        D[i, 0] = (x[i] ** 2 + y[i] ** 2) ** 0.5
        D[i, 1] = np.arctan(x[i] / y[i])
    return D
    
def generate_z(x0, y0, Vx, Vy, T, N, sigma_eta_D, sigma_eta_b):
    X = generate_X(x0, y0, Vx, Vy, T, N)
    D = generate_D(X[:, 0], X[:, 2])
    sigma_eta_D, sigma_eta_b = 20, 0.02
    eta = generate_eta(sigma_eta_D, sigma_eta_b, N)
    Dm = D + eta
    Zc = trans_to_cart(Dm)
    return Zc, Dm, generate_R(Dm, sigma_eta_D, sigma_eta_b)

def generate_eta(sigma_eta_D, sigma_eta_b, N):
    eta = np.zeros([N, 2])
    eta[:, 0] = np.random.normal(0, sigma_eta_D, N)
    eta[:, 1] = np.random.normal(0, sigma_eta_b, N)
    return eta

def generate_Dm_2T(D, sigma_eta_D, sigma_eta_b):
    eta = generate_eta(sigma_eta_D, sigma_eta_b, len(D) // 2)
    arr = np.zeros([len(eta), 2])
    for i in range(len(eta)):
        arr[i] = D[2 * i] + eta[i]
    return arr

def generate_b_T(D, sigma_eta_b1):
    eta = np.random.normal(0, sigma_eta_b1, len(D) // 2)
    arr = np.zeros(len(eta))
    for i in range(len(eta)):
        arr[i] = D[2 * i + 1, 1] + eta[i]
    return arr
        
def deriv_even(Xip):
    xip1 = Xip[0]
    yip1 = Xip[2]
    return np.array([[xip1 / (xip1 ** 2 + yip1 ** 2) ** 0.5, 0, yip1 / (xip1 ** 2 + yip1 ** 2) ** 0.5, 0],
                     [yip1 / (xip1 ** 2 + yip1 ** 2)       , 0, -xip1 / (xip1 ** 2 + yip1 ** 2)      , 0]])
    
def deriv_odd(Xip):
    xip1 = Xip[0]
    yip1 = Xip[2]
    return np.array([yip1 / (xip1 ** 2 + yip1 ** 2)       , 0, -xip1 / (xip1 ** 2 + yip1 ** 2)      , 0])

def h_even(X):
    x = X[0]
    y = X[2]
    return np.array([(x ** 2 + y ** 2) ** 0.5, np.arctan(x / y)])

def h_odd(X):
    x = X[0]
    y = X[2]
    return np.arctan(x / y)

def kalman_filt(z, zb, X0, P0, F, Q, R_even, R_odd):
    Xf = np.zeros([len(z), len(P0)])
    Xp = np.zeros([len(z), len(P0)])
    Pf = np.zeros([len(z), len(P0), len(P0)])
    Pp = np.zeros([len(z), len(P0), len(P0)])
    K = np.zeros([len(z), len(P0), len(z[0])])
    
    Xf_b = np.zeros([len(zb), len(P0)])
    Xp_b = np.zeros([len(zb), len(P0)])
    Pf_b = np.zeros([len(zb), len(P0), len(P0)])
    Pp_b = np.zeros([len(zb), len(P0), len(P0)])
    K_b = np.zeros([len(zb), len(P0)])
    
    Xf[0] = X0
    Xp[0] = X0
    Pf[0] = P0
    Pp[0] = P0
    for i in range(len(z) - 1):

        Xp[i + 1] = F.dot(Xf[i])
        Pp[i + 1] = (F.dot(Pf[i])).dot(F.T) + Q

        K = (Pp[i + 1].dot(deriv_even(Xp[i + 1]).T)).dot(np.linalg.inv((deriv_even(Xp[i + 1]).dot(Pp[i + 1])).dot(deriv_even(Xp[i + 1]).T) + R_even))

        Xf[i + 1] = Xp[i + 1] + K.dot(z[i + 1] - h_even(Xp[i + 1]))
        Pf[i + 1] = (np.identity(len(P0)) - K.dot(deriv_even(Xp[i + 1]))).dot(Pp[i + 1])
        
        Xf_b[i] = Xf[i]
                
        Xp_b[i + 1] = F.dot(Xf_b[i])
        Pp_b[i + 1] = (F.dot(Pf_b[i])).dot(F.T) + Q

        K_b = (Pp_b[i + 1].dot(deriv_odd(Xp_b[i + 1]).T)).dot(((deriv_odd(Xp_b[i + 1]).dot(Pp_b[i + 1])).dot(deriv_odd(Xp_b[i + 1]).T) + R_odd) ** (-1))

        Xf_b[i + 1] = Xp_b[i + 1] + K_b.dot(zb[i + 1] - h_odd(Xp_b[i + 1]))
        Pf_b[i + 1] = (np.identity(len(P0)) - K_b.dot(deriv_odd(Xp_b[i + 1]))).dot(Pp_b[i + 1])
        
        Xf[i] = Xf_b[i]
        

    return Xf, generate_D(Xf[:, 0], Xf[:, 2]), generate_D(Xp[:, 0], Xp[:, 2]), Pf, Pp

def kalman_filt_d(z, X0, P0, F, Q, R_even, R_odd):
    Xf = np.zeros([len(z), len(P0)])
    Xp = np.zeros([len(z), len(P0)])
    Pf = np.zeros([len(z), len(P0), len(P0)])
    Pp = np.zeros([len(z), len(P0), len(P0)])

    K = np.zeros([len(z), len(P0), len(z[0])])
    
    K_b = np.zeros([len(z), len(P0)])
    
    Xf[0] = X0
    Xp[0] = X0
    Pf[0] = P0
    Pp[0] = P0
    for i in range(len(z) - 1):
        if i % 2 == 1:
            Xp[i + 1] = F.dot(Xf[i])
            Pp[i + 1] = (F.dot(Pf[i])).dot(F.T) + Q

            K[i] = (Pp[i + 1].dot(deriv_even(Xp[i + 1]).T)).dot(np.linalg.inv((deriv_even(Xp[i + 1]).dot(Pp[i + 1])).dot(deriv_even(Xp[i + 1]).T) + R_even))

            Xf[i + 1] = Xp[i + 1] + K[i].dot(z[i + 1] - h_even(Xp[i + 1]))
            Pf[i + 1] = (np.identity(len(P0)) - K[i].dot(deriv_even(Xp[i + 1]))).dot(Pp[i + 1])
        else:
            
            Xp[i + 1] = F.dot(Xf[i])
            Pp[i + 1] = (F.dot(Pf[i])).dot(F.T) + Q

            K_b[i + 1] = (Pp[i + 1].dot(deriv_odd(Xp[i + 1]).T)).dot(((deriv_odd(Xp[i + 1]).dot(Pp[i + 1])).dot(deriv_odd(Xp[i + 1]).T) + R_odd) ** (-1))

            Xf[i + 1] = Xp[i + 1] + K_b[i].dot(z[i + 1, 1] - h_odd(Xp[i + 1]))
            Pf[i + 1] = (np.identity(len(P0)) - K_b[i].dot(deriv_odd(Xp[i + 1]))).dot(Pp[i + 1])
        
    return Xf, generate_D(Xf[:, 0], Xf[:, 2]), generate_D(Xp[:, 0], Xp[:, 2]), Pf, Pp

def prob(exp, var, T, x0, vx0, y0, vy0, N, sigma_eta_D, sigma_eta_b, X0, P0, F, Q):
    
    err_D = np.zeros([500, N - 3, 2])
    err_D_pred = np.zeros([500, N - 3, 2])
    for i in range(500):
        
        X = generate_trajectory(exp, var, T, x0, vx0, y0, vy0, N)
        D = generate_D(X[:, 0], X[:, 2])
        
        eta = generate_eta(sigma_eta_D, sigma_eta_b, N)
        Dm = D + eta
        
        R = np.array([[sigma_D_2, 0],
                      [0, sigma_b_2]])
        
        X_filt = kalman_filt(Dm, X0, P0, F, Q, R)
        
        err_D[i, :, 0] = np.array(D[3:, 0] - X_filt[1][3:, 0]) ** 2
        err_D[i, :, 1] = np.array(D[3:, 1] - X_filt[1][3:, 1]) ** 2
        err_D_pred[i, :, 0] = np.array(D[3:, 0] - X_filt[2][3:, 0]) ** 2
        err_D_pred[i, :, 1] = np.array(D[3:, 1] - X_filt[2][3:, 1]) ** 2
        
    final_err = np.zeros([N - 3, 4])
    for i in range(N - 3):
        final_err[i] = np.array([(np.mean(err_D[:, i, 0])) ** 0.5, (np.mean(err_D[:, i, 1])) ** 0.5,
                                (np.mean(err_D_pred[:, i, 0])) ** 0.5, (np.mean(err_D_pred[:, i, 1])) ** 0.5])
    return final_err

def prob_d(exp, var, T, x0, vx0, y0, vy0, N, sigma_eta_D, sigma_eta_b, sigma_eta_b1, P0, F, Q):
    
    err_D = np.zeros([500, N - 3, 2])
    err_D_pred = np.zeros([500, N - 3, 2])
            
    R_even = np.array([[sigma_eta_D ** 2, 0],
                       [0, sigma_eta_b ** 2]])
        
    R_odd = sigma_eta_b1 ** 2
    
    
    
    
    
    for i in range(500):
        X = generate_trajectory(0, sigmaa_2, T, x0, vx0, y0, vy0, N)
        
        D = generate_D(X[:, 0], X[:, 2])
        
        Dm_2T = generate_Dm_2T(D, sigma_eta_D, sigma_eta_b)
        
        bm = generate_b_T(D, sigma_eta_b1)
        
        Xm = trans_to_cart(Dm_2T)
        
        X0 = np.array([Xm[0, 0]                      ,
                      (Xm[0, 1] - Xm[0, 0]) / (2 * T),
                       Xm[1, 0]                      ,
                      (Xm[1, 1] - Xm[1, 0]) / (2 * T)])
        
        Dm = np.zeros([len(D), 2])
        for i in range(len(Dm_2T)):
            Dm[2 * i] = Dm_2T[i]
            Dm[2 * i + 1, 1] = bm[i]
#             print(Dm)
        
        X_filt_d = kalman_filt_d(Dm, X0, P0, F, Q, R_even, R_odd)
        
        err_D[i, :, 0] = np.array(D[3:, 0] - X_filt_d[1][3:, 0]) ** 2
        err_D[i, :, 1] = np.array(D[3:, 1] - X_filt_d[1][3:, 1]) ** 2
        err_D_pred[i, :, 0] = np.array(D[3:, 0] - X_filt_d[2][3:, 0]) ** 2
        err_D_pred[i, :, 1] = np.array(D[3:, 1] - X_filt_d[2][3:, 1]) ** 2
        
    final_err = np.zeros([N - 3, 4])
    for i in range(N - 3):
        final_err[i] = np.array([(np.mean(err_D[:, i, 0])) ** 0.5, (np.mean(err_D[:, i, 1])) ** 0.5,
                                (np.mean(err_D_pred[:, i, 0])) ** 0.5, (np.mean(err_D_pred[:, i, 1])) ** 0.5])
    return final_err

def trans_to_cart(Dm):
    Zc = np.zeros([len(Dm), 2])
    Zc[:, 0] = Dm[:, 0] * np.sin(Dm[:, 1])
    Zc[:, 1] = Dm[:, 0] * np.cos(Dm[:, 1])
    return Zc
```
